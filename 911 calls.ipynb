There is a column called title which is the title for the reason people called 911. Let us create a new column called reason and put the reasons in that column from the title column


```python
x= calls['title']
```

```python
calls['Reason'] = calls['title'].apply(lambda title: title.split(':')[0])
```

```python
type(calls['timeStamp'][0])
```

This is a string datatype. We need to convert this into Datatime type. 

```python
calls['timeStamp'] = pd.to_datetime(calls['timeStamp'])
```

```python
time =calls['timeStamp']
```

```python
calls['year']= time.apply(lambda time: time.year)
calls['month']= time.apply(lambda time: time.month)
calls['day'] = time.apply(lambda time: time.day)
calls['hour']= time.apply(lambda time: time.hour)
calls['dayofweek']= time.apply(lambda time: time.dayofweek)
calls['Date']= time.apply(lambda time: time.date())
```

```python
dict = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
calls['dayofweek'] = calls['dayofweek'].map(dict)
```

```python
# Now let us see on which day most of the calls based on the reason happened
```

```python
sns.countplot(x='dayofweek',hue='Reason',data=calls)
```

```python
sns.countplot(x='month',hue='Reason',data=calls)
```

Now let us see the count of calls per month

```python
month = calls.groupby('month').count()
```

```python
calls.groupby('Date').count()['lat'].plot()
```

We can say that we got more calls between Jan and February based on the spike in the graph

```python
# Now let us recreate the plots with the specific reason for the calls. So we will get three different plots 
```

```python
calls[calls['Reason']== 'Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
```

Traffic calls are more in the month of Jan

```python
calls[calls['Reason']== 'Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
```

Fire calls are more in the month of March 2016

```python
calls[calls['Reason']== 'EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
```

EMS calls are significantly very less during June 2016.
